# Price and Area Cleaning functions


```python
def parse_price_simple(price_str):
    """
    Convert text like:
      'PKR\\n4.75 Crore'
      'PKR\\n65 Lakh'
      'PKR\\n80 Thousand'
    into numeric PKR values using:
        1 Crore   = 10,000,000
        1 Lakh    =   100,000
        1 Thousand =   1,000
    """
    if pd.isna(price_str):
        return np.nan


    # Convert to string and clean basic stuff
    s = str(price_str)
    s = s.replace("PKR", "").replace("\n", " ").strip()

    # Split by space, e.g. "4.75 Crore" -> ["4.75", "Crore"]
    parts = s.split()

    if len(parts) == 0:
        return np.nan

    # First part is the number (remove commas)
    num_str = parts[0].replace(",", "")
    try:
        num = float(num_str)
    except ValueError:
        return np.nan

    # If there is a unit (Crore / Lakh / Thousand)
    if len(parts) > 1:
        unit = parts[1].lower()
    else:
        unit = ""

    if "crore" in unit:
        return num * 1e7
    elif "lakh" in unit:
        return num * 1e5
    elif "thousand" in unit:
        return num * 1e3
    else:
        # If no unit or unknown, just return the numeric part
        return num


def parse_area_simple(area_str):
    """
    Convert area values with units into square feet using:
        1 Marla  â‰ˆ 225 sq.ft.
        1 Kanal  = 20 Marla = 4,500 sq.ft.
        1 sq.yd. = 9 sq.ft.
    Examples:
        '128 Sq. Yd.' -> 128 * 9
        '5 Marla'     -> 5 * 225
        '1 Kanal'     -> 1 * 4500
        '1000 Sq. Ft.' -> 1000
    """
    if pd.isna(area_str):
        return np.nan

    s = str(area_str).strip()
    parts = s.split()

    if len(parts) == 0:
        return np.nan

    # First part is numeric
    num_str = parts[0].replace(",", "")
    try:
        num = float(num_str)
    except ValueError:
        return np.nan

    # Everything after the number is considered "unit text"
    unit_text = " ".join(parts[1:]).lower()

    # Check different unit possibilities
    if "marla" in unit_text:
        return num * 225.0
    elif "kanal" in unit_text:
        return num * 4500.0
    elif "sq" in unit_text:
        # square yards or square feet
        if "yd" in unit_text:
            return num * 9.0   # 1 sq.yd = 9 sq.ft
        else:
            return num         # assume already sq.ft
    else:
        # Unknown unit, return NaN to be safe
        return np.nan
```

# Apply price and area cleaning and create features

```python

df["Price_PKR"] = df["Price"].apply(parse_price_simple)
df["Area_SqFt"] = df["Area"].apply(parse_area_simple)

# Price per square foot (main comparison metric)
df["Price_per_SqFt"] = df["Price_PKR"] / df["Area_SqFt"]

# Quick sanity check
df[["Price", "Price_PKR", "Area", "Area_SqFt", "Price_per_SqFt"]].head()
```

# Standardize City Names

```python
# Strip leading/trailing spaces from City
df["City_clean"] = df["City"].astype(str).str.strip()

print("Number of unique raw City values:", df["City"].nunique())
print("Number of unique cleaned City values:", df["City_clean"].nunique())

df["City_clean"].value_counts().head(10)
```

```python
unique_cities = df["City_clean"].dropna().unique().tolist()

def fuzzy_fix_city(city_name):
    """
    Very light fuzzy example just to show concept.
    """
    if pd.isna(city_name):
        return city_name
    best_match, score = process.extractOne(city_name, unique_cities)
    return best_match

# Don't overwrite everything; just show a couple of examples
example_list = df["City_clean"].dropna().unique()[:5]
for c in example_list:
    print(c, "->", fuzzy_fix_city(c))
```

# Fix Bedrooms & Bathrooms Placeholders

```python
for col in ["Bedrooms", "Bathrooms"]:
    if col in df.columns:
        # Replace "-" with proper NaN
        df[col] = df[col].replace("-", np.nan)

        # Convert to numeric (non-numeric will become NaN)
        df[col] = pd.to_numeric(df[col], errors="coerce")

df[["Bedrooms", "Bathrooms"]].head()
```

# Filter For-Sale Listings & Remove Price Outliers

```python
# Keep only clear "For Sale" records
df_sale = df[df["Purpose"] == "For Sale"].copy()

print("For Sale listings (before cleaning):", df_sale.shape)

# Remove rows where essential fields are missing or invalid
df_sale = df_sale[(df_sale["Price_PKR"] > 0) & (df_sale["Area_SqFt"] > 0)]
df_sale["Price_per_SqFt"] = df_sale["Price_PKR"] / df_sale["Area_SqFt"]

print("For Sale listings (after dropping bad price/area):", df_sale.shape)

# IQR method for outlier detection on Price_PKR
Q1 = df_sale["Price_PKR"].quantile(0.25)
Q3 = df_sale["Price_PKR"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Upper bound for price outliers:", upper_bound)

# Keep only non-outlier prices
df_sale_clean = df_sale[df_sale["Price_PKR"] <= upper_bound].copy()

print("For Sale listings (after outlier removal):", df_sale_clean.shape)
```
